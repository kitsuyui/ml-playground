```python
import sympy
sympy.init_printing()
import sympy.abc as abc

x = abc.x
y = abc.y
a = abc.a
b = abc.b
c = abc.c
proportional = sympy.Eq(y, a * x)  # y = ax
linear = sympy.Eq(y, a * x + b)  # y = ax + b
quadratic = sympy.Eq(y, a * x ** 2 + b * x + c)  # y = ax^2 + bx + c
display(proportional)
display(linear)
display(quadratic)
```

```python
import sympy.plotting
eq1 = proportional.rhs.subs({a: 2})  # y = 2x
eq2 = linear.rhs.subs({a: 2, b: 3})  # y = 2x + 3
eq3 = quadratic.rhs.subs({a: 2, b: 3, c: 4})  # y = 2x^2 + 3x + 4
display(eq1, eq2, eq3)
p1 = sympy.plotting.plot(eq1, show=False)
p2 = sympy.plotting.plot(linear.rhs.subs({a: 2, b: 3}), show=False)
p3 = sympy.plotting.plot(quadratic.rhs.subs({a: 2, b: 3, c: 4}), show=False)
p1.extend(p2)
p1.extend(p3)
p1.show()
```
