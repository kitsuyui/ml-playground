```python
import numpy as np
import matplotlib.pyplot as plt


def log_return(x: np.ndarray) -> np.ndarray:
    """Compute the log return of a time series."""
    return np.diff(np.log(x))


# ja: log_return は対数収益率を計算する関数。
# 対数収益率は、もとは投資や金融工学で使われる概念。
# (収益) = (金額 t2) - (金額 t1)
# (収益率) = (金額 t2) / (金額 t1)
# (対数収益率) = log(金額 t2) - log(金額 t1)
# 収益率を考えるメリット: もとの価格の単位やボリュームに依存しない効果を比較できる
#  (例: 1000円 -> 2000円 は 2倍の収益率、2000ドル -> 4000ドル は 2倍の収益率)
# 対数収益率を考えるメリット: 収益率のメリットに加え、計算・管理が簡単になること (対数法則)
# (対数収益率 t2-t1) = log(金額 t2) - log(金額 t1) = log(金額 t2 / 金額 t1)
# (対数収益率 t3-t2) = log(金額 t3) - log(金額 t2) = log(金額 t3 / 金額 t2)
# があるとき、 (対数収益率 t3-t1) = log(金額 t3 / 金額 t1) = (対数収益率 t3-t2) + (対数収益率 t2-t1) となる。
# このように、対数収益率は足し算・引き算だけで計算できる。
# 通常の収益率に戻すときは exp(対数収益率) とすればよい。

# en: log_return is a function that computes the log return of a time series.
# The log return is a concept used in investment and financial engineering.
# (return) = (price t2) - (price t1)
# (return rate) = (price t2) / (price t1)
# (log return) = log(price t2) - log(price t1)
# The advantage of return rate: we can compare the effect that is independent of the unit and volume of the original price.
# (e.g. 1000 yen -> 2000 yen is 2 times the return rate, 2000 dollars -> 4000 dollars is 2 times the return rate)
# The advantage of log return: in addition to the advantage of return rate, it is easy to calculate and manage (logarithmic law)
# (log return t2-t1) = log(price t2) - log(price t1) = log(price t2 / price t1)
# (log return t3-t2) = log(price t3) - log(price t2) = log(price t3 / price t2)
# When there is (log return t3-t1) = log(price t3 / price t1) = (log return t3-t2) + (log return t2-t1),
# the log return can be calculated by addition and subtraction.
# To return to the normal return rate, just use exp(log return).


example_data_1 = np.power(np.linspace(1, 2, 1000), 2) * 50 + 200
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(example_data_1, 'g')
ax2.plot(log_return(example_data_1), 'b')
ax1.set_xlabel('time')
ax1.set_ylabel('price', color='g')
ax2.set_ylabel('logreturn', color='b')


# sine wave
example_data_2 = np.sin(np.linspace(0, 2 * np.pi, 1000)) * 50 + 200
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(example_data_2, 'g')
ax2.plot(log_return(example_data_2), 'b')
ax1.set_xlabel('time')
ax1.set_ylabel('price', color='g')
ax2.set_ylabel('logreturn', color='b')
```
